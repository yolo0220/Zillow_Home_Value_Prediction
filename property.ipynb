```python
def corrmatrix(df):
    corrmat = df.corr()
    plt.subplots(figsize=(11, 9))
    cmap = sns.diverging_palette(220, 10, as_cmap=True)
    sns.heatmap(corrmat, vmax=.8, square=True, cmap=cmap)
    
corrmatrix(data2)
```

```python
def outlier1(data, feature):
    low = np.percentile(data[feature], 2.5)
    up = np.percentile(data[feature], 97.5)

    data = data[(data[feature] > low) & (data[feature] < up)]
    return data

data3 = outlier1(data2, "price")
```

```python
data3 = outlier1(data3, "cfsf")
```

```python
model = sm.OLS.from_formula("price ~ bath + bed + qualitytype + cfsf + aircondition + heating + spa + pool + C(landuse) + year", data=data3).fit()
print(model.summary())

influence = model.get_influence()
cooks_d2, pvals = influence.cooks_distance
fox_cr = 4 / (len(data3) - model.df_model - 1)
idx = np.where(cooks_d2 > fox_cr)[0]
```

```python
a = np.arange(len(data3))
data3 = data3.set_index(a)
```

```python
out_data = data3.iloc[idx,]
```

```python
#독립변수들 표준화 
from sklearn.preprocessing import StandardScaler
def std(data):
    numeric_columns = ["aircondition", "bath", "bed", "cfsf", "spa", "pool", "year"]
    stdData = StandardScaler().fit_transform(data[numeric_columns])
    stdData = pd.DataFrame(stdData, columns=numeric_columns)
    return stdData
```

```python
from sklearn.decomposition import PCA as sklearnPCA
def pca(data):
    tempPCA = sklearnPCA(n_components=5)
    newIndex = tempPCA.fit_transform(data)
    newData = pd.DataFrame(newIndex)
    return newData, tempPCA
```

```python
stdData = std(data2)
stdData.head()
```
